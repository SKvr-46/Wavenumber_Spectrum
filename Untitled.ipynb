```python
omegas =  np.array(omegas)+(-1j * 2 * np.pi / 1000)
def u_values_omega_domain(omegas):
    u_values = []
    for omega in omegas:
        u_sum = 0j
        for n in n_range:
            k = 2 * n * np.pi / L
            if np.real(omega) >0:    #虚部が負のものを選択
                nu = np.sqrt((omega / alpha) ** 2 - k ** 2)
            else:
                nu = -1 * np.sqrt((omega / alpha) ** 2 - k ** 2)
            if(np.imag(nu) > 0):
                print("this is error")
            if n == 0:
                u_sum += (1j * np.pi / L) * (k ** 2 / nu) * scipy.special.jv(1, k * r) * np.exp(-1j * nu * np.abs(z))
            else:
                u_sum += (2 * (1j * np.pi / L) * (k ** 2 / nu) * scipy.special.jv(1, k * r) * np.exp(-1j * nu * np.abs(z)))
        u_values.append(u_sum)
    return np.array(u_values)
```

```python
import numpy as np

def DFT(f_t):
  # 結果を格納するリスト
  F_omega = []

  # データ長
  length = f_t.shape[0]

  # データ長だけ繰り返します
  for omega in range(length):

    # x[n]に掛け合わせる部分です
    weight = np.exp(-1j*((2*np.pi*omega)/length))

    # シグマの部分です
    F = 0
    for t in range(length):
        F += f_t[t] * (weight**t)
    F_omega.append(F)
  return np.array(F_omega)

# サンプリング周波数
fs = 100
# サンプリング周期
ts = 1.0/fs
# 時間軸
t = np.arange(0, 1, ts)
# 周波数5Hzのサイン波
f = np.sin(2*np.pi*5*t)

F = DFT(f)


# プロット
plt.plot(np.abs(F), label='DFT')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

plt.show()
```

```python
def IDFT(F_omega):
  # 結果を格納するリスト
  f_t = []

  # データ長
  length = F_omega.shape[0]

  # データ長だけ繰り返します
  for t in range(length):

    # X[k]に掛け合わせる部分です
    weight = np.exp(1j*((2*np.pi*t)/length))

    # シグマの部分です
    f = 0
    for omega in range(length):
        f += F_omega[omega] * (weight**omega)
    f_t.append(f/length)  # 逆変換では和をデータ長で割ります
  return np.array(f_t)

f = IDFT(F)

# プロット
plt.plot(t, f, label='original')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Define the time range and sampling rate
t_start = 0  # start time
t_end = 10  # end time
sampling_rate = 1000  # sampling rate in Hz
t_fft = np.linspace(t_start, t_end, int((t_end - t_start) * sampling_rate), endpoint=False)

# Define the function for the FFT
sin_t_fft = np.sin(np.pi * t_fft)

# Perform the FFT
fft_sin_t = fft(sin_t_fft)

# Define the frequency range
freqs = fftfreq(len(t_fft), 1/sampling_rate)

# We only plot the positive frequencies, as the FFT is symmetric around 0
positive_freqs = freqs[:len(freqs)//2]
fft_sin_t_positive = fft_sin_t[:len(freqs)//2]

# Plot the FFT
plt.figure(figsize=(12, 6))

# Plot the magnitude spectrum
plt.subplot(1, 2, 1)
plt.plot(positive_freqs, np.abs(fft_sin_t_positive))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Magnitude spectrum of sin(πt)')

# Plot the phase spectrum
plt.subplot(1, 2, 2)
plt.plot(positive_freqs, np.angle(fft_sin_t_positive))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (radians)')
plt.title('Phase spectrum of sin(πt)')

plt.tight_layout()
plt.show()
```
