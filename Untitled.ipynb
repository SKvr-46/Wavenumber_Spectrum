```python
import numpy as np
import scipy.special
from scipy.fft import fft, ifft, fftfreq

L = 25 * 10**3
alpha = 5 * 10**3
r = 10 * 10**3
z = 1 * 10**3
n_range = np.arange(0, 100)
m_range = np.arange(0, 100)
omegas = np.linspace(-200, 200, 800)
t = np.linspace(0, 10, len(omegas))


#u(omega)
def u_values_omega_domain(omegas, n_range, m_range, L, alpha, r, z):
    u_values = []
    for omega in omegas:
        u_sum = 0
        for n in n_range:
            for m in m_range:
                k_n = 2 * n * np.pi / L
                k_m = 2 * m * np.pi / L
                nu = round(np.sqrt((omega/alpha)**2 - k_n**2 - k_m**2), 4) - (1e-6j) if ((omega/alpha)**2 - k_n**2 - k_m**2) > 0 else -1j * np.sqrt( k_n**2 + k_m**2- (omega/alpha)**2) - (1e-6j)
                if n == 0:
                    u_sum += (-2* np.pi / L**2) * (k_n / nu) * np.exp(-1j * k_n * r) * np.exp(-1j * nu * np.abs(z))
                else:
                    u_sum += (2 * (-2 * np.pi / L**2) * (k_n / nu) * np.exp(-1j * k_n * r) * np.exp(-1j * nu * np.abs(z)))
        u_values.append(u_sum)
    return np.array(u_values)

#u*f
def convolution_integral(t, omegas, n_range, m_range, L, alpha, r, z):
    # Compute u_values in frequency domain
    u_omega = u_values_omega_domain(omegas, n_range, m_range, L, alpha, r, z)
    
    u_t =  ifft(u_omega)
    
    # Define f(t)
    f_t = 1/2 * (1 + np.tanh(t / 0.125))
    
    conv_integral = ifft(u_omega * fft(f_t))
    
    return conv_integral

plt.plot(t, convolution_integral(t, omegas, n_range,m_range, L, alpha, r, z))
plt.xlabel('Time (t)')
plt.ylabel('Displacement (u)')
plt.title('Displacement in the Time Domain')
plt.grid(True)
plt.show()
```
