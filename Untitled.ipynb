```python
import numpy as np
import scipy.special
import matplotlib.pyplot as plt

L = 25 * 10**3
alpha = 5 * 10**3
r = 10 * 10**3
z = 1 * 10**3
n_range = np.arange(0, 1000)
omegas = np.linspace(0, 6000, 20000)
t = np.array(np.linspace(0, 30, len(omegas)))
dt = t[1] - t[0]
omega_imag = (-1j * 2 * np.pi / 10 ** 6)
omegas = np.array(omegas + omega_imag)#w = w - λi
```

```python
nu_values = []
def u_values_omega_domain(omega, n_range, L , alpha, r, z):
    u_values = []
    for omega in omegas:
        u_sum = 0
        for n in n_range:
            k = 2 * n * np.pi / L
            nu = np.sqrt((omega / alpha) ** 2 - k ** 2)
            nu_values.append(nu) #確認用(Im(nu)<0の確認)
            if n == 0:
                u_sum += ((1j * np.pi / L) * (k ** 2 / nu) * scipy.special.jv(1, k * r) * np.exp(-1j * nu * np.abs(z)))
            else:
                u_sum += (2 * (1j * np.pi / L) * (k ** 2 / nu) * scipy.special.jv(1, k * r) * np.exp(-1j * nu * np.abs(z)))
        u_values.append(u_sum)
    return np.array(u_values)
```

```python
u_omega = u_values_omega_domain(omegas, n_range, L, alpha, r, z)
```

```python
domega = omegas[1] - omegas[0]
```

```python
u_t = []
for ti in t:
    u_ti = np.sum(u_omega * np.exp(1j * omegas * ti) * domega)
    u_t.append(u_ti.real)

def f_t_values(t):
    return 1 / 2 * (1 + np.tanh(t / 0.125))

f_t = f_t_values(t)

# Compute the convolution integral in the time domain directly
conv_integral = np.convolve(f_t, u_t, 'same') * dt

# Plotting the convolution integral
plt.plot(t, conv_integral)
plt.xlabel('Time (t)')
plt.ylabel('Displacement (u)')
plt.title('Convolution Integral in Time Domain')
plt.grid(True)
plt.xlim(0, 10)
plt.show()
```

```python
import numpy as np
import scipy.special
import matplotlib.pyplot as plt

L = 25
alpha = 5
r = 10
z = 1
n_range = np.arange(0, 1000)
omegas = np.linspace(0,6000, 4000)
t = np.linspace(0, 10, len(omegas))
dt = t[1] - t[0]

omega_imag = (-1j * 2 * np.pi / 10 ** 6)
omegas = omegas + omega_imag

def u_values_omega_domain(omegas, n_range, L, alpha, r, z):
    u_values = []
    for omega in omegas:
        u_sum = 0
        for n in n_range:
            k = 2 * n * np.pi / L
            nu = np.sqrt((omega / alpha) ** 2 - k ** 2)
            if n == 0:
                u_sum += (1j * np.pi / L) * (k ** 2 / nu) * scipy.special.jv(1, k * r) * np.exp(-1j * nu * np.abs(z))
            else:
                u_sum += (2 * (1j * np.pi / L) * (k ** 2 / nu) * scipy.special.jv(1, k * r) * np.exp(-1j * nu * np.abs(z)))
        u_values.append(u_sum)
    return np.array(u_values)

u_omega = u_values_omega_domain(omegas, n_range, L, alpha, r, z)

# Fourier transform
u_t = np.fft.ifft(u_omega)

def f_t_values(t):
    return 1 / 2 * (1 + np.tanh(t / 0.125))

f_t = f_t_values(t)

# Compute the convolution integral in the time domain directly
conv_integral = np.convolve(f_t, u_t, 'same') * dt

# Apply Hamming window to u_t
window = np.hamming(len(conv_integral ))
conv_u_t_windowed = np.array(conv_integral ) * window


# Plotting the convolution integral
plt.plot(t, conv_u_t_windowed)
plt.xlabel('Time (t)')
plt.ylabel('Displacement (u)')
plt.title('Convolution Integral in Time Domain')
plt.grid(True)
plt.xlim(0,10)
plt.show()
```

```python
import numpy as np
import scipy.special
import matplotlib.pyplot as plt

L = 25
alpha = 5
r = 10
z = 1
n_range = np.arange(0, 1000)
t = np.linspace(0, 10, 10000)
dt = (t[1] - t[0]) / 1000
N = len(t)
omegas =  np.linspace(0, 5000, N) # Shift the zero frequency to the center
omega_imag = (-1j * 2 * np.pi / 10 ** 6)
omegas = omegas + omega_imag

def dft(f_t):
    n = len(f_t)
    f_omega = []
    for x in range(n):
        y = 0j
        for t in range(n):
            a = 2 * np.pi * t * x / n + omega_imag * t
            y += f_t[t] * np.exp(-1j * a)
        f_omega.append(y)
    return f_omega

def dft_(f_omega):
    n = len(f_omega)
    f_t = []
    for x in range(n):
        y = 0j
        for t in range(n):
            a = 2 * np.pi * t * x / n - omega_imag * t
            y += f_omega[t] * np.exp(1j * a)
        f_t.append(y)
    return f_t


def u_values_omega_domain(omegas, n_range, L, alpha, r, z):
    u_values = []
    for omega in omegas:
        u_sum = 0
        for n in n_range:
            k = 2 * n * np.pi / L
            nu = np.sqrt((omega / alpha) ** 2 - k ** 2)
            if n == 0:
                u_sum += (1j * np.pi / L) * (k ** 2 / nu) * scipy.special.jv(1, k * r) * np.exp(-1j * nu * np.abs(z))
            else:
                u_sum += (2 * (1j * np.pi / L) * (k ** 2 / nu) * scipy.special.jv(1, k * r) * np.exp(-1j * nu * np.abs(z)))
        u_values.append(u_sum)
    return np.array(u_values)

u_omega = u_values_omega_domain(omegas, n_range, L, alpha, r, z)

# Fourier transform
#u_t = np.fft.ifft(u_omega)  # Shift the zero frequency to the center of u_omega before the inverse FFT
u_t = dft(u_omega)


def f_t_values(t):
    return 1 / 2 * (1 + np.tanh(t / 0.125))

f_t = f_t_values(t)

# Compute the convolution integral in the time domain directly
conv_integral = np.convolve(f_t, u_t, 'same') * dt

# Apply Hamming window to u_t
window = np.hamming(len(conv_integral))
conv_u_t_windowed = np.array(conv_integral) * window

# Plotting the convolution integral
plt.plot(t, conv_u_t_windowed)
plt.xlabel('Time (t)')
plt.ylabel('Displacement (u)')
plt.title('Convolution Integral in Time Domain')
plt.grid(True)
plt.xlim(0,10)
plt.show()
```
