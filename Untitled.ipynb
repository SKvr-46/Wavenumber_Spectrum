## TRY DISCRETE FOURIER TRANSFORM WITH REVERSE FOURIER TRANSEFORM
### wの虚部は正になる。時間と共に減少する項を入れてiwtと合わせることで、それを得る。

##### 直すべきところ：tは実際に定義しているところと、フーリエ変換部分のとで異なるので、横軸が期待通りでない。
##### tを(19)の条件内で調べてみる

```python
L = 25
alpha = 5
r = 10
z = 1

n_range = np.arange(0, 1000)
t = np.linspace(0, 20, 10000)
dt = (np.abs(t[1] - t[0])) / 1000
N = len(t)
omega_imag = (-1j * 2 * np.pi / 20)
```

```python
#フーリエ変換:f(t) => F(ω)
def DFT(f_t):
  # 結果を格納するリスト
  F_omega = []

  # データ長
  length = f_t.shape[0]

  # データ長だけ繰り返します
  for omega in range(length):

    # x[n]に掛け合わせる部分です
    weight = np.exp(-1j*((2*np.pi*omega)/length))

    # シグマの部分です
    F = 0
    for t in range(length):
        F += f_t[t] * (weight**t)
    F_omega.append(F/length)
  return np.array(F_omega)
```

```python
#フーリエ(逆)変換:F(ω)=> f(t) 
def IDFT(F_omega):
  # 結果を格納するリスト
    f_t = []
    length = F_omega.shape[0]

    for t in range(length):

    # X[k]に掛け合わせる部分です
        weight = np.exp(1j*((2*np.pi*t)/length))

    # シグマの部分です
        f = 0
        for omega in range(length):
            f += F_omega[omega] * (weight**(omega + omega_imag)) #プラスをかける
        f_t.append(f)  # 逆変換では和をデータ長で割ります
    return np.array(f_t)
```

```python
omegas =  np.array(range(N)) + omega_imag   #care N = len(u_omega)
def u_values_omega_domain(omegas, n_range, L, alpha, r, z):
    u_values = []
    for omega in omegas:
        u_sum = 0
        for n in n_range:
            k = 2 * n * np.pi / L
            nu = np.sqrt((omega / alpha) ** 2 - k ** 2)    #-1をかけて、虚部が負のものを選択
            if(np.imag(nu) > 0):
                print("this is error")
            if n == 0:
                u_sum += (1j * np.pi / L) * (k ** 2 / nu) * scipy.special.jv(1, k * r) * np.exp(-1j * nu * np.abs(z))
            else:
                u_sum += (2 * (1j * np.pi / L) * (k ** 2 / nu) * scipy.special.jv(1, k * r) * np.exp(-1j * nu * np.abs(z)))
        u_values.append(u_sum)
    return np.array(u_values)
```

```python
omegas =  np.array(range(N)) + omega_imag   #care N = len(u_omega)
def phi_values_omega_domain(omegas, n_range, L, alpha, r, z):
    u_values = []
    for omega in omegas:
        u_sum = 0
        for n in n_range:
            k = 2 * n * np.pi / L
            nu = np.sqrt((omega / alpha) ** 2 - k ** 2)    #-1をかけて、虚部が負のものを選択
            if(np.imag(nu) > 0):
                print("this is error")
            if n == 0:
                u_sum += (-1j * np.pi / L) * (k / nu) * scipy.special.jv(0, k * r) * np.exp(-1j * nu * np.abs(z))
            else:
                u_sum += (2 * (-1j * np.pi / L) * (k / nu) * scipy.special.jv(0, k * r) * np.exp(-1j * nu * np.abs(z)))
        u_values.append(u_sum)
    return np.array(u_values)
```

```python
#エラー：
x_set = np.linspace(0,100,1000)
n_set = np.arange(0,10)

def y_val(x_set,n_set):
    y_values = []
    for n in n_set:
        u_sum = 0
        for x in x_set:
            u_sum += n * x
        y_values.append(u_sum)
    return np.array(y_values)

y = y_val(x_set, n_set)
plt.plot(x_set, y, label='original')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```
