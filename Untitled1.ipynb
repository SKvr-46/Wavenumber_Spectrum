## DISCRETE WAVE NUMBER METHOD

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.special
import math
```

```python
### (離散化) ###
L = 25
alpha = 5
r = 10
z = 1

n_range = np.arange(0, 1000)
omega_imag = (-1j * 2 * np.pi)
```

```python
f_s = 500 # サンプリングレート f_s[Hz] (任意)
t_fin = 20 # 収録終了時刻 [s] (任意)
dt = 1/f_s # サンプリング周期 dt[s]
N = int(f_s * t_fin) # サンプル数 [個]
t = np.arange(0, t_fin, dt) 
omegas = np.fft.fftfreq(N, d=dt) # 周波数を割り当てる（※後述）
```

```python
def u_values_omega_domain(omegas):
    u_values = []
    for omega in omegas:
        u_sum = 0j
        for n in n_range:
            k = 2 * n * np.pi / L
            nu = np.sqrt((omega / alpha) ** 2 - k ** 2) if np.imag(np.sqrt((omega / alpha) ** 2 - k ** 2)) < 0 else  -1*np.sqrt((omega / alpha) ** 2 - k ** 2)
            if(np.imag(nu) > 0):
                print("this is error")
                break
            if n == 0:
                u_sum += (1j * np.pi / L) * (k ** 2 / nu) * scipy.special.jv(1, k * r) * np.exp(-1j * nu * np.abs(z))
            else:
                u_sum += (2 * (1j * np.pi / L) * (k ** 2 / nu) * scipy.special.jv(1, k * r) * np.exp(-1j * nu * np.abs(z)))
        u_values.append(u_sum)
    return np.array(u_values)
```

```python
def phi_values_omega_domain(omegas):
    u_values = []
    for omega in omegas:
        u_sum = 0j
        for n in n_range:
            k = 2 * n * np.pi / L
            nu = np.sqrt((omega / alpha) ** 2 - k ** 2) if np.imag(np.sqrt((omega / alpha) ** 2 - k ** 2)) < 0 else  -1*np.sqrt((omega / alpha) ** 2 - k ** 2)
            if(np.imag(nu) > 0):
                print("this is error")
                break
            if n == 0:
                u_sum += (-1j * np.pi / L) * (k / nu) * scipy.special.jv(0, k * r) * np.exp(-1j * nu * np.abs(z))
            else:
                u_sum += (2 * (-1j * np.pi / L) * (k / nu) * scipy.special.jv(0, k * r) * np.exp(-1j * nu * np.abs(z)))
        u_values.append(u_sum)
    return np.array(u_values)
```

## ここでやろうとしてること
1秒間に44100個のサンプルをとり、0から1秒までの収録を行う
この時、yは44100個の長さの配列になっている。y_fftで離散フーリエ変換し、
plt.plot(freq, Amp)で表示するこの時以下の注意が必要。

##### freqはnp.fft.fftfreq(N, d=dt)で定義しているように、44100個の周波数配列。これと周波数-振幅スペクトルの図を書く時、周波数は-22050から22050の範囲で書かれる。

ifftではこれを元に戻す。

```python
import numpy as np # numpyをインポート
import matplotlib.pyplot as plt # matplotlibをインポート

### 量子化(離散化) ###
f_s = 44100 # サンプリングレート f_s[Hz] (任意)
t_fin = 1 # 収録終了時刻 [s] (任意)
dt = 1/f_s # サンプリング周期 dt[s]
N = int(f_s * t_fin) # サンプル数 [個]

### 入力信号 y(t) ###
f0 = 440 # 周波数 f0[Hz]
t = np.arange(0, t_fin, dt) # 離散時間 t[s]
y = np.sin(2*np.pi*f0*t) # y(t) を生成

### y(t) のグラフ ###
plt.plot(t, y) # y-t グラフのプロット
plt.xlim(0, 10*10**-3) # 横軸に関する描画範囲指定
plt.show() # グラフの表示
```

```python
### FFT: tの関数をfの関数にする ###
y_fft = np.fft.fft(y) # 離散フーリエ変換
freq = np.fft.fftfreq(N, d=dt) # 周波数を割り当てる（※後述）
Amp = abs(y_fft/(N/2)) # 音の大きさ（振幅の大きさ）
```

```python
### 音波のスペクトル ###
#plt.plot(freq[1:int(N/2)], Amp[1:int(N/2)]) # A-f グラフのプロット
plt.plot(freq, Amp) # A-f グラフのプロット
#plt.xscale("log") # 横軸を対数軸にセット
plt.show()
```

```python
### IFFT: fの関数をtの関数にする ###
y_ifft = np.fft.ifft(y_fft)
```

```python
### 逆変換したグラフ: もとの波形を復元 ###
plt.plot(t, y_ifft.real) # 実部しかいらない
plt.xlim(0, 10*10**-3)
plt.show()
```

```python
import numpy as np

def ifft_by_hand(frequency_domain, frequency_values, time_values):
    result = np.zeros_like(time_values, dtype=complex)
    for i, t in enumerate(time_values):
        for omega, y in zip(frequency_values, frequency_domain):
            result[i] += y * np.exp(1j * omega * t) *np.exp(np.abs(omega_imag) * t)
    return result
```

```python
# Calculate u_values in omega domain
omegas = np.linspace(-1000, 1000, 2002)  # You can adjust these parameters as needed
omegas =  np.array(omegas)+omega_imag
```

```python
calc_u_values_omega_domain = u_values_omega_domain(omegas)  # Assuming you have this function

# Convert to time domain
time_values = np.linspace(0, 20, 2002)  # You can adjust these parameters as needed
calc_u_values_time_domain = ifft_by_hand(calc_u_values_omega_domain, omegas, time_values)
```

```python
calc_u_values_omega_domain = phi_values_omega_domain(omegas)  # Assuming you have this function

# Convert to time domain
time_values = np.linspace(0, 20, 2002)  # You can adjust these parameters as needed
calc_u_values_time_domain = ifft_by_hand(calc_u_values_omega_domain, omegas, time_values)
```

```python

plt.plot(time_values, calc_u_values_time_domain)
```

```python
# Define f_t function
def f_t_values(t):
    return 1 / 2 * (1 + np.tanh(t / 0.125))

# Calculate f_t values
f_t = f_t_values(time_values)

# Ensure that f_t and calc_u_values_time_domain are the same length
if len(f_t) != len(calc_u_values_time_domain):
    print("Lengths of f_t and calc_u_values_time_domain do not match!")
else:
    # Perform convolution
    convolved_values = np.convolve(f_t, calc_u_values_time_domain, mode='same')

plt.plot(time_values, convolved_values)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the frequency_domain values for 1/(i*pi*omega)
# Avoid zero frequency to prevent division by zero
omegas = np.linspace(-1000, 1000, 2001)  # You can adjust these parameters as needed
omegas = omegas[omegas != 0]  # exclude zero
frequency_domain = 1 / (1j * np.pi * omegas)

# Inverse Fourier transform
time_values = np.linspace(-10, 10, 2001)  # You can adjust these parameters as needed
time_domain = ifft_by_hand(frequency_domain, omegas, time_values)

# Plot the result
plt.figure(figsize=(10, 6))
plt.plot(time_values, time_domain.real, label='Real part')
plt.plot(time_values, time_domain.imag, label='Imaginary part')
plt.legend()
plt.title("Inverse Fourier Transform of 1/(i*pi*omega)")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```
