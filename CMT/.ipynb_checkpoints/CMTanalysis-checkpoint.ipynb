```python
import os
import csv
import pandas as pd

def txt_to_csv(directory):
    for filename in os.listdir(directory):
        if filename.endswith("CMTLIST.txt"):
            txt_filepath = os.path.join(directory, filename)
            csv_filepath = os.path.join(directory, filename.replace("CMTLIST.txt", "CMTLIST.csv"))

            with open(txt_filepath, "r") as txt_file, open(csv_filepath, "w", newline="") as csv_file:
                writer = csv.writer(csv_file)
                for line in txt_file:
                    writer.writerow(line.strip().split())

# ディレクトリパスを指定して関数を呼び出す例
txt_to_csv("/Users/koei/Desktop/CMT")
```

```python
csv_filepath = "/Users/koei/Desktop/CMT/CMTLIST.csv"
df = pd.read_csv(csv_filepath)
```

```python
dep = list(df.iloc[:, 8])
```

```python
dip1 = list(df.iloc[:, 11])
```

```python
dip2 = list(df.iloc[:, 14])
```

```python
rak1 = list(df.iloc[:, 12])
```

```python
rak2 = list(df.iloc[:, 15])
```

```python
#浅いdepの、indexのみ取り出す
index_list = []
for i, d in enumerate(dep):
    if (d <= 30):
        index_list.append(i)
# index_list 
# for i in index_list:
#     print(dep[i])
```

```python
#dep<=30以下に対応するdipとrakeを取り出す。
opt_dip1 = []
opt_dip2 = []
opt_rak1 = []
opt_rak2 = []

for i in index_list:
    opt_dip1.append(dip1[i])
    opt_dip2.append(dip2[i])
    opt_rak1.append(rak1[i])
    opt_rak2.append(rak2[i])
```

```python
#もとのdepを30以下のみの配列に上書き。
to_remove = []
for d in dep:
    if d > 30:
        to_remove.append(d)  # 削除する要素を一時的にリストに追加

for item in to_remove:
    dep.remove(item)  # リストから要素を削除
```

```python
opt_summation_dip = []
for i in range(34078):
    opt_summation_dip.append(opt_dip1[i]+opt_dip2[i])
opt_summation_dip
avr_summation = sum(opt_summation_dip)/len(opt_summation_dip)
print(avr_summation)
```

```python
#正断層と逆断層のrake角の条件から、正断層と逆断層のrake角配列を作成。同時に対応するindexも格納する
normal_fault_rake_dip1 = []
normal_fault_rake_index_dip1 = []
reverse_fault_rake_dip1 = []
reverse_fault_rake_index_dip1 = []

for i, angle in enumerate(opt_rak1):
    if (-100 <= angle <= -80):
        normal_fault_rake_dip1.append(angle)
        normal_fault_rake_index_dip1.append(i)
    elif (80 <= angle <= 100):
        reverse_fault_rake_dip1.append(angle)
        reverse_fault_rake_index_dip1.append(i)
```

```python
#dep30以下の場合のdipから、正断層に当たるもののみ取り出す
normal_fault_dip_dip1 = []
for index in normal_fault_rake_index_dip1:
    normal_fault_dip_dip1.append(opt_dip1[index])
    
#dep30以下の場合のdipから、逆断層に当たるもののみ取り出す
reverse_fault_dip_dip1 = []
for index in reverse_fault_rake_index_dip1:
    reverse_fault_dip_dip1.append(opt_dip1[index])
```
