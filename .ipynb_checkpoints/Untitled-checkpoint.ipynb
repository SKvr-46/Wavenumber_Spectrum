```python
わかったこと
速度が最大になる点(r/beta)が、ヘビサイドのように常に1であれば、Mの倍率がないので、きれいにr^2で減衰するが、
tanh(x)のように、立ち上がりが遅いと、r/betaで1ではない値が取れる。このときの倍率も減衰の倍率に加味されるので、綺麗な−２乗の減衰がみられない。

ただなぜtanh(x)のときも変位は良い結果をみせているかというと、変位は変位の最大点を使っている
一方で、速度は、速度が最大になる点でのM0が1になっているとは限らない点にある。
```

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# パラメータ
alpha = 5
beta = 3
r_values = [10, 20, 50, 100]

# 畳み込む関数
def f(t):
    return np.heaviside(t, 0)

def g(t, r):
    return t * (np.heaviside(t - r / alpha, 0) - np.heaviside(t - r / beta, 0)) / r ** 4

# tの値
t = np.linspace(0, 50, 10000)
dt = t[1] - t[0]

# 畳み込み積分用の範囲
t_full = np.linspace(0, 100, len(t) * 2 - 1)

# subplot設定
fig, axs = plt.subplots(3, 1, figsize=(12, 18))

for r in r_values:
    # f,gを微分する
    f_prime_t = np.gradient(f(t), dt)
    g_prime_t = np.gradient(g(t, r), dt)

    # f*g 畳み込み
    conv = signal.convolve(f(t), g(t, r), mode='full') * dt

    # 畳み込みの微分(f*g)'
    d_conv = np.gradient(conv, dt)

    # Plot 
    axs[0].plot(t_full, d_conv, label=f"(f*g)' for r={r}")

    # Plot 
    axs[1].plot(t_full, conv, label=f"f*g for r={r}")

    # Plot both in the new subplot
    axs[2].plot(t_full, d_conv, label=f"(f*g)' for r={r}")
    axs[2].plot(t_full, conv, label=f"f*g for r={r}")

# Titles and labels
axs[0].set_title("(f*g)' for different r values")
axs[0].set_xlabel('t')
axs[0].set_ylabel('km/s')

axs[1].set_title("convolution for different r values")
axs[1].set_xlabel('t')
axs[1].set_ylabel('km')

axs[2].set_title("Both (f*g)' and f*g for different r values")
axs[2].set_xlabel('t')
axs[2].set_ylabel('km/s and km')

# Add legends and titles
for ax in axs:
    ax.legend()
    ax.grid(True)
    ax.set_xlim([0, 50])
    ax.set_ylim([0, 0.0004])
    ax.set_xticks(np.arange(0, 50, 3.3))

plt.tight_layout()
plt.show()
```
