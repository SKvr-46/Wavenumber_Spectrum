```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate_u(n_values, x, t_values):
    u_values = []
    for t in t_values:
        u_sum = 0
        for n in n_values:
            u = 2 / (n * np.pi) * np.sin(n * np.pi) * np.sin(n * np.pi * x) * np.sin(n * np.pi * t)
            u_sum += u
        u_values.append(u_sum)
    return u_values

n_values = range(1, 31)
x = 1
t_values = np.linspace(0, 100, 500)

u_values = calculate_u(n_values, x, t_values)

plt.plot(t_values, u_values)
plt.xlabel('t')
plt.ylabel('u')
plt.title('Graph of u with x = 1')
plt.show()
```
